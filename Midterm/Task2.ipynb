```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
import matplotlib.image as pltimg

im = pltimg.imread('minion.jpg')
hist = cv2.calcHist([im],[0],None,[256],[0,256])

fig = plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(im)
plt.subplot(122)
plt.imshow(hist)
plt.show()
```
